```python
m = 1
l = 5
a = 30*np.pi/180
k = np.linspace(1,10,100)
theta0 = np.linspace(-10,-1,100)
K,Theta0 = np.meshgrid(k,theta0)
thetadot = np.sqrt((-K/(m*l**2)*4*a*Theta0)/(np.cos(2*a)**-2-1))

theta01 = (2*m*l*9.8*(1-np.cos(-a)) - k*a**2)/(-2*a*k)
thetadot2 = np.sqrt((-k/(m*l**2)*4*a*theta0)/(np.cos(2*a)**-2-1))
theta02 = (2-l/9.8*thetadot2**2-3*np.cos(-a)-k*a**2/(m*9.8*l))*(m*9.8*l/(k*2*a))

fig = plt.figure(1, figsize=(8,8))
plt.imshow(thetadot,aspect='auto', origin='lower', extent=(K.min(),K.max(),Theta0.min(),Theta0.max()))
plt.plot(k,theta01,'r')
plt.plot(k,theta02,'g')
plt.title('Steadystate walking speed heatmap')
plt.xlabel('k')
plt.ylabel('theta0')
plt.axis([K.min(),K.max(),Theta0.min(),Theta0.max()])
plt.legend(['Cannot start from rest (above line)','Will loose contact with the ground (below line)'])
```
